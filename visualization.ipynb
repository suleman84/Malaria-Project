```
parapath = "gdrive/My Drive/cell-imgs/Parasitized/"
uninpath = "gdrive/My Drive/cell-imgs/Uninfected/"
parastized = os.listdir(parapath)
uninfected = os.listdir(uninpath)
```

```
plt.figure(figsize = (12,12))
for i in range(6):
    plt.subplot(1, 6, i+1)
    img = cv2.imread('gdrive/My Drive/cell-imgs/Parasitized' + "/" + parastized[i+10])
    plt.imshow(img)
    plt.title('PARASITIZED : ' + str(i+1))
    plt.tight_layout()
plt.show()
```
